# Bivariate

#### Person age vs laon status

- Age ranging from 22 to 26 is generally considered to be associated with lower credit risk compared to other age groups.

```python
grouped_data = pd.crosstab(df['Age Group'], df['loan_status'])

# Plot the grouped bar plot
ax = grouped_data.plot(kind='bar', stacked=False, figsize=(10, 6))

# Set labels and title
plt.title('Age Group vs Loan Status')
plt.xlabel('Loan Grade')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend(title='Loan Status')

# Calculate total counts for each loan grade
total_counts_by_grade = grouped_data.sum(axis=1)

# Calculate percentages and add annotations
for i in range(len(grouped_data.index)):
    for j in range(len(grouped_data.columns)):
        count = grouped_data.iloc[i, j]
        percentage = (count / total_counts_by_grade[i]) * 100
        ax.text(
            i + j * 0.2 - 0.15, count + 0.5,  # Adjust the position of the text
            f'{percentage:.1f}%', 
            ha='center', va='bottom'
        )

# Remove grid lines for better readability
# ax.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

#### cb_person_default_on_file vs loan_status

- loan-intent 'no' has highest count for loan status 0 

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='cb_person_default_on_file', hue='loan_status', data=df)

# Set labels and title
plt.xlabel('cb person default on file')
plt.ylabel('Count')
plt.title('cb person default on file vs Loan Status')

# Calculate total counts for each 'cb_person_default_on_file' category
total_counts = df.groupby('cb_person_default_on_file')['loan_status'].count()

# Calculate percentages and add annotations
for p in ax.patches:
    height = p.get_height()
    # Get the 'cb_person_default_on_file' category for the current bar
    cb_person_default_on_file = p.get_x() + p.get_width() / 2.0
    # Adjust the index to ensure correct category mapping
    cb_person_default_on_file = 'Y' if round(cb_person_default_on_file) == 0 else 'N'
    
    # Get the total count for this category
    total = total_counts[cb_person_default_on_file]
    
    # Calculate percentage
    if height != 0:  # Skip annotating bars with zero count
        percentage = (height / total) * 100
        
        # Annotate the bar with the percentage
        ax.text(p.get_x() + p.get_width() / 2., height + 0.5, f'{percentage:.1f}%', ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```
