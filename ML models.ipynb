```python
X_train.dtypes
```

### Separate Numerical and Categorical Features

```python
X_train_cat_num = X_train[['person_home_ownership','loan_intent','cb_person_default_on_file','Age Group','Income Group','Loan amnt Group']]

X_train_cat_order= X_train[['loan_grade']]
X_train_num = X_train[['person_emp_length','loan_int_rate','loan_percent_income','cb_person_cred_hist_length']]
```

### Numerical Feature Transformation 
* Normalization
* Standardization (We will use this method in this project)
### Categorical Feature Transformation 
* One Hot Encoding
* Label Encoding
* Ordinal Encoding (We will us this method in this project)

### Tranformation on X_train Data

### Numerical Feature Transformation: Applying Standardization

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

X_train_num_transformed = pd.DataFrame(std_scaler.fit_transform(X_train_num), 
                                    columns = std_scaler.get_feature_names_out(), 
                                    index = X_train_num.index)

X_train_num_transformed.head()
```

### Categorical Feature Transformation: Applying Ordinal Encoding(order)

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

X_train_cat_ordinal_transformed = X_train_cat_order.apply(lambda col: label_encoder.fit_transform(col))

X_train_cat_ordinal_transformed.columns = X_train_cat_order.columns

X_train_cat_ordinal_transformed.head()
```

### Categorical Feature Transformation: Applying One hot encoding(nominal)

```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder(sparse_output=False, 
                               handle_unknown="ignore")

# column names are (annoyingly) lost after OneHotEncoding
# (i.e. the dataframe is converted to a numpy ndarray)
X_train_cat_onehot_tansformed = pd.DataFrame(onehot_encoder.fit_transform(X_train_cat_num), 
                               columns=onehot_encoder.get_feature_names_out(), 
                               index = X_train_cat_num.index)

# print("Shape of Data before Transformation:", X_train_cat.shape)
# print("Shape of Data after Transformation:", X_train_cat_tansformed.shape)

X_train_cat_onehot_tansformed.head()
```

### Concatenate X_train_num_transformed and X_train_cat_transformed

```python
X_train_transformed = pd.concat([X_train_num_transformed, X_train_cat_onehot_tansformed,X_train_cat_ordinal_transformed], axis=1)

X_train_transformed.head()
```

### Step 7 - Apply Data Preparation on X_test Data

```python
# X_test_cat = X_test.select_dtypes(include=['object'])
# X_test_num = X_test.select_dtypes(include=['int64', 'float64'])
X_test_cat_num = X_test[['person_home_ownership','loan_intent','cb_person_default_on_file','Age Group','Income Group','Loan amnt Group']]

X_test_cat_order= X_test[['loan_grade']]
X_test_num = X_test[['person_emp_length','loan_int_rate','loan_percent_income','cb_person_cred_hist_length']]
```

### Numerical Feature Transformation: Applying Standardization


```python
X_test_num_transformed = pd.DataFrame(std_scaler.transform(X_test_num), 
                                   columns = std_scaler.get_feature_names_out(), 
                                   index = X_test_num.index)

X_test_num_transformed.head()
```

### Categorical Feature Transformation: Applying Ordinal Encoding


```python
label_encoders = {}
for column in X_train_cat_order.columns:
    label_encoders[column] = LabelEncoder()
    label_encoders[column].fit(X_train_cat_order[column])

# Apply label encoding to each column in X_test_cat_order
X_test_cat_ordinal_transformed = X_test_cat_order.copy()
for column in X_test_cat_order.columns:
    X_test_cat_ordinal_transformed[column] = label_encoders[column].transform(X_test_cat_order[column])

# Display the transformed DataFrame
X_test_cat_ordinal_transformed.head()
```

### Categorical Feature Transformation: Applying One hot encoding(nominal columns)


```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder(sparse_output=False, 
                               handle_unknown="ignore")

# column names are (annoyingly) lost after OneHotEncoding
# (i.e. the dataframe is converted to a numpy ndarray)
X_test_cat_onehot_tansformed = pd.DataFrame(onehot_encoder.fit_transform(X_test_cat_num), 
                               columns=onehot_encoder.get_feature_names_out(), 
                               index = X_test_cat_num.index)

# print("Shape of Data before Transformation:", X_train_cat.shape)
# print("Shape of Data after Transformation:", X_train_cat_tansformed.shape)

X_test_cat_onehot_tansformed.head()
```

### Concatenate X_test_num_transformed and X_test_cat_transformed

```python
X_test_transformed = pd.concat([X_test_num_transformed,X_test_cat_onehot_tansformed, X_test_cat_ordinal_transformed], axis=1)

X_test_transformed.head()
```

```python
from sklearn import metrics
```

### KNN Classifier

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import classification_report
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
df.info()
```

```python
from tqdm import tqdm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in X_train.columns:
    if X_train[column].dtype == 'object':  # Check if the column is categorical
        label_encoders[column] = LabelEncoder()
        X_train[column] = label_encoders[column].fit_transform(X_train[column])

# Apply the same encoding to the test data (X_test)
for column in X_test.columns:
    if column in label_encoders:
        X_test[column] = label_encoders[column].transform(X_test[column])

# Initialize GridSearchCV
tuned_parameters = [{'n_neighbors': [i for i in range(1, 51)], 
                     'p': [1, 2, 3]}]

clf = RandomizedSearchCV(
    estimator=KNeighborsClassifier(), 
    param_distributions=tuned_parameters, 
    scoring='f1',
    cv=5,
    return_train_score=True,
    verbose=1
)
clf.fit(X_train_transformed, y_train)
```

```python
print("Best parameters set found on train set:")
print(clf.best_params_)
print(clf.best_estimator_)
```

```python
model_filename = 'knn_classifier_model.joblib'
joblib.dump(clf, model_filename)
print(f"Model saved to {model_filename}")
```

```python
knn=load('knn_classifier_model.joblib')
y_pred = knn.predict(X_test_transformed)

# Evaluate the model
f1 = metrics.f1_score(y_test, y_pred)
print(f'f1_score: {f1:.2f}')

# Additional metrics
print(metrics.classification_report(y_test, y_pred))
print(metrics.confusion_matrix(y_test, y_pred))
```

### Decision Tree

```python
from tqdm import tqdm
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np

label_encoders = {}
for column in X_train.columns:
    if X_train[column].dtype == 'object':  # Check if the column is categorical
        label_encoders[column] = LabelEncoder()
        X_train[column] = label_encoders[column].fit_transform(X_train[column])

# Apply the same encoding to the test data (X_test)
for column in X_test.columns:
    if column in label_encoders:
        X_test[column] = label_encoders[column].transform(X_test[column])

# Initialize RandomizedSearchCV with DecisionTreeClassifier
tuned_parameters = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None] + list(np.arange(1, 21)),
    'min_samples_split': np.arange(2, 11),
    'min_samples_leaf': np.arange(1, 11),
    'max_features': [None, 'sqrt', 'log2']  # Removed 'auto'
}

clf = RandomizedSearchCV(
    estimator=DecisionTreeClassifier(), 
    param_distributions=tuned_parameters, 
    scoring='f1',
    cv=5,
    n_iter=100,  # Number of parameter settings sampled
    return_train_score=True,
    verbose=1,
    random_state=42  # Ensuring reproducibility
)
clf.fit(X_train_transformed,y_train)
```

```python
# Get the best parameters and best score
print(f"Best parameters: {clf.best_params_}")
print(f"Best score: {clf.best_score_}")
```

```python
model_filename = 'dt_classifier_model.joblib'
joblib.dump(clf, model_filename)
print(f"Model saved to {model_filename}")
```

```python
dt=load('dt_classifier_model.joblib')
y_pred = dt.predict(X_test_transformed)

# Evaluate the model
f1 = metrics.f1_score(y_test, y_pred)
print(f'f1 score: {f1:.2f}')

# Additional metrics
print(metrics.classification_report(y_test, y_pred))
print(metrics.confusion_matrix(y_test, y_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import RandomizedSearchCV
from tqdm import tqdm
import numpy as np
import joblib
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Define hyperparameters for RandomizedSearchCV
param_distributions = {
    'n_estimators': [int(x) for x in np.linspace(start=50, stop=500, num=10)],  # Number of trees in the forest
    'max_features': [None, 'sqrt', 'log2'],  # Number of features to consider at each split
    'max_depth': [int(x) for x in np.linspace(10, 100, num=10)] + [None],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]  # Minimum number of samples required to be at a leaf node
}

# Initialize RandomizedSearchCV
clf = RandomizedSearchCV(
    estimator=RandomForestClassifier(),
    param_distributions=param_distributions,
    n_iter=50,  # Number of parameter settings to sample
    scoring='f1',  # Use 'f1' instead of 'f1_score'
    cv=5,
    return_train_score=True,
    verbose=1,  # Set to 1 to use internal verbosity
    random_state=42  # For reproducibility
)
clf.fit(X_train_transformed, y_train)
```

```python
# Get the best parameters and best score
print(f"Best parameters: {clf.best_params_}")
print(f"Best score: {clf.best_score_}")
```

```python
model_filename = 'rf_model.joblib'
joblib.dump(clf, model_filename)
print(f"Model saved to {model_filename}")
```

```python
rf=load('rf_model.joblib')
y_pred = rf.predict(X_test_transformed)

# Evaluate the model
f1 = metrics.f1_score(y_test, y_pred)
print(f'f1 score: {f1:.2f}')

# Additional metrics
print(metrics.classification_report(y_test, y_pred))
print(metrics.confusion_matrix(y_test, y_pred))
```
